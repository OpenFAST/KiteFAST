```python
#set a few things before running the analysis

#column location of roll pitch yaw rotation in KiteFAST.out
roll_col=44
pitch_col=45
yaw_col=46

#column location of Fx,FY,FZ (global integrated aero loads) in the KiteFAST.out
kf_fx_col=47
kf_fy_col=48
kf_fz_col=49

kf_surge_col=19
kf_sway_col=20
kf_heave_col=21
kf_roll_col=22
kf_pitch_col=23
kf_yaw_col=24

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial.transform import Rotation as R
import matplotlib.pyplot as plt


#load the KiteFAST output file
KF_pd=pd.read_csv('KiteFast.out',skiprows=7,delim_whitespace=True)
KF=KF_pd.values

```

```python
#plot the aero loads in global coordinates
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]
plt.plot(KF[:,0], KF[:,kf_surge_col],label='surge KiteFAST')
plt.plot(KF[:,0], KF[:,kf_sway_col],label='sway KiteFAST')
plt.plot(KF[:,0], KF[:,kf_heave_col],label='heave KiteFAST')
plt.legend()
plt.title('Platform Translation')
plt.ylabel('Position [m]')
plt.xlabel('Time [s]')
plt.xlim([0.0, 60.0])
#plt.ylim([-300000, 100000])
plt.show()
```

```python
#plot the aero loads in global coordinates
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]
plt.plot(KF[:,0], KF[:,kf_roll_col],label='roll KiteFAST')
plt.plot(KF[:,0], KF[:,kf_pitch_col],label='pitch KiteFAST')
plt.plot(KF[:,0], KF[:,kf_yaw_col],label='yaw KiteFAST')
plt.legend()
plt.title('Platform Rotation')
plt.ylabel('Rotation [deg]')
plt.xlabel('Time [s]')
plt.xlim([0.0, 60.0])
#plt.ylim([-300000, 100000])
plt.show()
```
